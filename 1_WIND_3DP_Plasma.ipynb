```python
import pandas as pd
import numpy as np

data_dict = {}

# ---------------------------------------------------------
# IF BLOCK: Handle SpacePy Data Model
# ---------------------------------------------------------
if 'spacepy' in str(type(data)):
    print("Detected SpacePy data. Processing...")
    for var in var_names:
        var_array = np.array(data[var])
        
        # Safely get attributes
        attrs = data[var].attrs if hasattr(data[var], 'attrs') else {}
        
        # Handle Fill Values
        if 'FILLVAL' in attrs:
            fill_val = attrs['FILLVAL']
            if var_array.dtype.kind in 'fc': 
                var_array = np.where(np.isclose(var_array, fill_val), np.nan, var_array)
            
        # --- 1D Arrays ---
        if len(var_array.shape) == 1:
            if 'Epoch' in data and len(var_array) == len(data['Epoch']):
                data_dict[var] = var_array
                
        # --- 2D Arrays (Vectors, Tensors) ---
        elif len(var_array.shape) == 2:
            num_components = var_array.shape[1]
            col_labels = []
            
            if 'LABL_PTR_1' in attrs:
                label_var_name = attrs['LABL_PTR_1']
                if label_var_name in data:
                    raw_labels = np.array(data[label_var_name])
                    for lbl in raw_labels:
                        clean_lbl = lbl.decode('utf-8') if isinstance(lbl, bytes) else str(lbl)
                        col_labels.append(clean_lbl.strip())
            
            for i in range(num_components):
                if len(col_labels) == num_components:
                    col_name = f"{var}_{col_labels[i]}"
                else:
                    col_name = f"{var}_{i}"
                    
                if 'Epoch' in data and len(var_array[:, i]) == len(data['Epoch']):
                    data_dict[col_name] = var_array[:, i]

# ---------------------------------------------------------
# ELSE BLOCK: Handle xarray Data Model (cdflib)
# ---------------------------------------------------------
else:
    print("Detected xarray data. Processing...")
    
    # In xarray, 'Epoch' is often a coordinate. Let's get its length safely.
    epoch_length = len(data['Epoch'].values) if 'Epoch' in data else 0
    
    for var in var_names:
        # Skip variables that might not be in data_vars (like coordinates)
        if var not in data.data_vars:
            continue
            
        # Extract the raw numpy array using .values
        var_array = data.data_vars[var].values
        
        # Safely get attributes
        attrs = data.data_vars[var].attrs
        
        # Handle Fill Values
        if 'FILLVAL' in attrs:
            fill_val = attrs['FILLVAL']
            if var_array.dtype.kind in 'fc': 
                var_array = np.where(np.isclose(var_array, fill_val), np.nan, var_array)
                
        # --- 1D Arrays ---
        if len(var_array.shape) == 1:
            if epoch_length > 0 and len(var_array) == epoch_length:
                data_dict[var] = var_array
                
        # --- 2D Arrays (Vectors, Tensors) ---
        elif len(var_array.shape) == 2:
            num_components = var_array.shape[1]
            col_labels = []
            
            if 'LABL_PTR_1' in attrs:
                label_var_name = attrs['LABL_PTR_1']
                
                # Fetch labels (can be in data_vars or coordinates in xarray)
                if label_var_name in data:
                    raw_labels = data[label_var_name].values
                    for lbl in raw_labels:
                        clean_lbl = lbl.decode('utf-8') if isinstance(lbl, bytes) else str(lbl)
                        col_labels.append(clean_lbl.strip())
            
            for i in range(num_components):
                if len(col_labels) == num_components:
                    col_name = f"{var}_{col_labels[i]}"
                else:
                    col_name = f"{var}_{i}"
                    
                if epoch_length > 0 and len(var_array[:, i]) == epoch_length:
                    data_dict[col_name] = var_array[:, i]

# ---------------------------------------------------------
# Create the DataFrame
# ---------------------------------------------------------
df = pd.DataFrame(data_dict)

if 'Epoch' in df.columns:
    df.set_index('Epoch', inplace=True)
elif 'Epoch' in data: # Fallback: explicitly add the Epoch from original data if missed
    if 'spacepy' in str(type(data)):
        df.index = np.array(data['Epoch'])
    else:
        df.index = data['Epoch'].values

print(f"\nDataFrame shape: {df.shape}")
print("Columns:", df.columns.tolist())
```
